# 01 Extraction - Loading the Data

This notebook validates that the raw dataset can be loaded correctly before cleaning.

What this notebook checks:
- The file path is correct and readable.
- Column names are usable (including fixing extra spaces in headers).
- Basic structure, data types, and missing values are visible.
- A raw snapshot is saved for traceability.

### Imports

Load core libraries used in this notebook:
- `pandas` for tabular data loading and inspection.
- `numpy` for numerical utilities (used in later notebooks and kept consistent).
- `os` for basic path/system operations if needed.

```python
import pandas as pd
import numpy as np
import os
```

### File Paths

Define the relative path to the raw CSV file.

Using a single variable for the path makes the notebook easier to maintain and reduces hard-coded path errors.

```python
RAW_PATH = "../data/raw/APY.csv"
```

### Load Dataset

Read the CSV into a DataFrame and normalize the column names.

Important detail:
- `df.columns = df.columns.str.strip()` removes extra spaces in headers (for example `District `), which prevents column lookup errors later.

`df.head()` is shown to quickly confirm the dataset loaded correctly.

```python
df = pd.read_csv(RAW_PATH)
df.columns = df.columns.str.strip()
df.head()
```

### Basic Structure

Inspect the dataset dimensions and column list.

This confirms:
- Number of rows and columns.
- Final column names after header cleanup.
- Whether the schema matches expectations before deeper analysis.

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
```

### Data Types

Use `df.info()` to inspect:
- Data type of each column.
- Non-null count per column.
- Memory usage summary.

This helps identify columns that may need type conversion during cleaning.

```python
df.info()
```

### Missing Values

Count null values in each column with `df.isnull().sum()`.

This provides a column-level null profile so we can decide what to drop, impute, or keep in the cleaning notebook.

```python
df.isnull().sum()
```

### Unique Values (Key Columns)

Check cardinality for core categorical fields (`State`, `District`, `Crop`, `Season`).

Why this matters:
- Confirms expected geographic and category coverage.
- Helps estimate grouping complexity for later EDA and KPI analysis.
- Can reveal unexpected duplicates or naming inconsistencies.

```python
for col in ["State", "District", "Crop", "Season"]:
    print(f"{col}: {df[col].nunique()}")
```

### Save Raw Snapshot

Export the current extracted DataFrame to `../data/processed/raw_snapshot.csv`.

This creates a stable checkpoint of the extracted data (with cleaned headers) so downstream notebooks can be compared and reproduced reliably.

```python
df.to_csv("../data/processed/raw_snapshot.csv", index=False)
```
